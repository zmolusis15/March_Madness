```python
import pandas as pd
```

```python
eff = pd.read_csv('efficiency.csv')
```

```python
eff.head()
```

```python
eff = eff[['Team','Conference','Tempo-Adj','Off. Efficiency-Adj', 
           'Def. Efficiency-Adj']]
```

```python
eff = eff.rename(columns = {'Team': 'team', 'Conference' : 'conference', 
                      'Tempo-Adj':'tempo_adj', 'Off. Efficiency-Adj':'off_eff_adj',
                     'Def. Efficiency-Adj':'def_eff_adj'})
```

```python
four_fact = pd.read_csv('fourFact.csv')
```

```python
four_fact.columns
```

```python
four_fact = four_fact[['Team','Off-eFG%','Off-TO%', 'Off-OR%', 'Off-FTRate',
                       'Def-eFG%','Def-TO%','Def-OR%','Def-FTRate']]
```

```python
df = pd.merge(left=eff, right=four_fact, left_on=['team'], right_on=['Team'], 
         how='outer') 
```

```python
df = df.drop(['Team'], axis = 1)
```

```python
exp = pd.read_csv('exp.csv')
```

```python
exp = exp.drop(['Unnamed: 0'],axis = 1)
```

```python
df = pd.merge(left=df, right=exp, left_on=['team'], right_on=['Team'], 
         how='outer') 
```

```python
df = df.drop(['Team'], axis = 1)
```

```python
ft_3pt = pd.read_csv('ft_3pt.csv')
```

```python
ft_3pt = ft_3pt.drop(['Unnamed: 0'],axis = 1)
```

```python
df = pd.merge(left=df, right=ft_3pt, left_on=['team'], right_on=['Team'], 
         how='outer') 
```

```python
df = df.drop(['Team'], axis = 1)
```

```python
df.columns
```

```python
sos = pd.read_csv('SOS_32.csv')
```

```python
df = pd.merge(left=df, right=sos, left_on=['team'], right_on=['Team'], 
         how='outer')
```

```python
df = df.drop(['Team'], axis = 1)
```

```python
df = df.set_index('team')
```

```python
df.to_csv('final_df.csv')
```
